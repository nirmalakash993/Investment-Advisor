we first create an example DataFrame named companies.We then define the conditions we want to filter the DataFrame by using a list of conditions. We use the len function to check if a condition causes blank rows. If a condition causes blank rows, we skip that condition using a conditional statement. We then combine the remaining conditions using the & operator to get the final set of conditions. We use these conditions to filter the DataFrame

```python
# initial filtering expression for High risk profile
HR = (companies['Market Cap(Cr)'] < 2000) & \
       (companies['10-Year Return(%)'] < 8) & \
       (companies['DELTA '] > 0)

# check if any condition gives zero rows
if companies[HR].shape[0] == 0:
    # remove the first condition and re-evaluate the expression
    HR = (companies['10-Year Return(%)'] < 8) & \
           (companies['DELTA '] > 0)
elif companies[HR].shape[0] == 0:
    # remove the second condition and re-evaluate the expression
    HR = (companies['Market Cap(Cr)'] < 2000) & \
           (companies['DELTA '] > 0)
elif companies[HR].shape[0] == 0:
    # remove the third condition and re-evaluate the expression
    HR = (companies['Market Cap(Cr)'] < 2000) & \
           (companies['10-Year Return(%)'] < 8)

# apply the final filtering expression
High_Risk = companies[HR]
High_Risk=High_Risk.sort_values('Dividend Per Share',ascending=False).head(5)

High_Risk_Comp=High_Risk['Company'].tolist()
```

```python
# initial filtering expression for risk TAKING profile
R = (companies['Market Cap(Cr)'].between(2000, 5000)) & \
       (companies['10-Year Return(%)'].between(8, 15)) & \
       (companies['DELTA '] > 0)


# check if any condition gives zero rows
if companies[R].shape[0] == 0:
    # remove the first condition and re-evaluate the expression
    R = (companies['10-Year Return(%)'].between(8, 15)) & \
           (companies['DELTA '] > 0)
elif companies[R].shape[0] == 0:
    # remove the second condition and re-evaluate the expression
    R = (companies['Market Cap(Cr)'].between(2000, 5000)) & \
           (companies['DELTA '] > 0)
elif companies[R].shape[0] == 0:
    # remove the third condition and re-evaluate the expression
    R = (companies['Market Cap(Cr)'].between(2000, 5000)) & \
           (companies['10-Year Return(%)'].between(8, 15))

# apply the final filtering expression
Risk = companies[R]
Risk=Risk.sort_values('Dividend Per Share',ascending=False).head(5)
Risk_Comp=Risk['Company'].tolist()
```

```python
# initial filtering expression for MODERATE risk profile
MR = (companies['Market Cap(Cr)'].between(5000, 15000)) & \
       (companies['10-Year Return(%)'].between(15, 20)) & \
       (companies['DELTA '] > 0)


# check if any condition gives zero rows
if companies[MR].shape[0] == 0:
    # remove the first condition and re-evaluate the expression
    MR = (companies['10-Year Return(%)'].between(15, 20)) & \
           (companies['DELTA '] > 0)
elif companies[MR].shape[0] == 0:
    # remove the second condition and re-evaluate the expression
    MR = (companies['Market Cap(Cr)'].between(5000, 15000)) & \
           (companies['DELTA '] > 0)
elif companies[MR].shape[0] == 0:
    # remove the third condition and re-evaluate the expression
    MR = (companies['Market Cap(Cr)'].between(5000, 15000)) & \
           (companies['10-Year Return(%)'].between(15, 20))

# apply the final filtering expression
Moderate_Risk = companies[MR]
Moderate_Risk=Moderate_Risk.sort_values('Dividend Per Share',ascending=False).head(5)

Moderate_Risk_Comp=Moderate_Risk['Company'].tolist() 
```

```python
# initial filtering expression for Low risk profile
LR = (companies['Market Cap(Cr)'] > 15000) & \
       (companies['10-Year Return(%)'] > 20) & \
       (companies['DELTA '] > 0)

# check if any condition gives zero rows
if companies[LR].shape[0] == 0:
    # remove the first condition and re-evaluate the expression
    LR = (companies['10-Year Return(%)'] > 20) & \
           (companies['DELTA '] > 0)
elif companies[LR].shape[0] == 0:
    # remove the second condition and re-evaluate the expression
    LR = (companies['Market Cap(Cr)'] > 15000) & \
           (companies['DELTA '] > 0)
elif companies[LR].shape[0] == 0:
    # remove the third condition and re-evaluate the expression
    LR = (companies['Market Cap(Cr)'] > 15000) & \
           (companies['10-Year Return(%)'] > 20)

# apply the final filtering expression
Low_Risk = companies[LR]
Low_Risk=Low_Risk.sort_values('Dividend Per Share',ascending=False).head(5)
Low_Risk_Comp=Low_Risk['Company'].tolist() 
```

```python
# Now its time to select companies according to Risk profile
```

```python
Best_company = [High_Risk_Comp,Risk_Comp,Moderate_Risk_Comp,Low_Risk_Comp]
prof=['High Risk Taking','Risk Taking','Moderate Risk Taking','Low Risk Taking']
cell_row = 2  # Start from row 2
for i in range(0, 4):
    for j in range(len(Best_company[i])):
        cell_value1 = Best_company[i][j]
        Best.update_cell(cell_row,2,prof[i])
        try:
            Best.update_cell(cell_row, 3, cell_value1)  # Column B
        except:
            pass
        cell_row = cell_row + 1
```
